# Customer Churn Analysis Notebook

This notebook performs customer churn analysis using synthetic data, including data preprocessing and visualizations.

```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from customer_churn_analysis import create_churn_dataframe

# Set up matplotlib for inline display
%matplotlib inline
plt.style.use('default')
```

```python
# Generate the dataset
data = create_churn_dataframe(num_rows=10000)
print("Dataset generated successfully!")
print(f"Dataset shape: {data.shape}")
data.head()
```

```python
# Check data info
data.info()
```

```python
# Data preprocessing
print("Null values:")
print(data.isnull().sum())

# Drop customer_id
data = data.drop('customer_id', axis=1)

# Map churn to 0/1
data['churn'] = data['churn'].map({'Yes': 1, 'No': 0})

print("Churn distribution:")
print(data['churn'].value_counts())

# Check for duplicates
print(f"Number of duplicates: {data.duplicated().sum()}")
data = data.drop_duplicates()
print(f"Dataset shape after removing duplicates: {data.shape}")

# One-hot encoding
data = pd.get_dummies(data, drop_first=True)
print("Columns after preprocessing:")
print(data.columns.tolist())
data.head()
```

```python
# Churn Distribution
plt.figure(figsize=(8, 6))
sns.countplot(x='churn', data=data)
plt.title('Churn Distribution')
plt.xlabel('Churn (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.show()
```

```python
# Churn by Contract (1 Year)
plt.figure(figsize=(8, 6))
sns.countplot(x='contract_1 year', hue='churn', data=data)
plt.title('Churn by Contract (1 Year)')
plt.xlabel('1 Year Contract (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.legend(title='Churn')
plt.show()
```

```python
# Churn by Gender
plt.figure(figsize=(8, 6))
sns.countplot(x='gender_Male', hue='churn', data=data)
plt.title('Churn by Gender')
plt.xlabel('Male (0 = Female, 1 = Male)')
plt.ylabel('Count')
plt.legend(title='Churn')
plt.show()

# Churn by Internet Service
plt.figure(figsize=(8, 6))
sns.countplot(x='internet_services_Fiber', hue='churn', data=data)
plt.title('Churn by Internet Service')
plt.xlabel('Fiber Internet (0 = DSL, 1 = Fiber)')
plt.ylabel('Count')
plt.legend(title='Churn')
plt.show()
```
